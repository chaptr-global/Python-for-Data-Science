# Matplotlib CheatSheet

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import ticker
plt.style.use('seaborn-white') #use plt.style.available[:] to find styles
%matplotlib inline
```

# Common Visulizations/Graphs

## Line Plots (Continuous Data)

```python
x = np.linspace(-np.pi, np.pi, 100)
y1 = cosine = np.cos(x)
y2 = sine = np.sin(x)
```

### Creating a Simple Line Plot

```python
plt.plot(x,y1) #cosine graph
plt.show()
```

### Creating Figures and Subplots

#### Method 1: Using plt.figure and plt.subplot(row,colum,panel number)

```python
plt.figure(figsize=(20,5)) #figsize to alter the width, height of plot

plt.subplot(2,1,1) #rows, columns, panel number
plt.plot(x,y1)

plt.subplot(2,1,2)
plt.plot(x,y2)
```

#### Method 2: Using plt.subplots(n) and axis indices

```python
fig, ax = plt.subplots(2, figsize=(20,5)) #create 2 subplots

ax[0].plot(x,y1,"r")
ax[1].plot(x,y2,"b")
```

### Making Plot Style Adjustments

#### Adding/Modifying Title, Labels, Legend, Limits, and Locators

```python
#Available Locators: Fixed, Linear, Null, Multiple, Index, Auto, MaxN, Log; Available formatters: Null, Index, Fixed, Func, FormatStr, Scalar, Log
fig,ax = plt.subplots(1, figsize=(20,5)) #create figure with only one subplot
ax.xaxis.set_major_locator(ticker.MultipleLocator(np.pi/2)) #set ticker locators for x-axis
## Change the ticker format to include pi, and add a secondary scale to include angle in degrees

#Available colors: rgbcmyk; available formats: "-" (solid),"-."(dash-dot),":"(dotted),"--"(dashed)
plt.plot(x,y1, "r--", label="cos(θ)") #xvalues, y-values, plot style (color/fmt), label (necessary for legend)
plt.plot(x,y2, "b", label="sin(θ)") 

plt.title("Cosine and Sine Graph", fontsize=20) #add plot title and adjust the font size
plt.xlim(-3.5,3.5) #set x-axis scale limits
plt.ylim(-1.25,1.25)

#ax.xaxis.set_major_locator(MultipleLocator(20))

plt.xlabel("θ (Radians)") #add x-axis label
plt.ylabel("cos/sin") #add y-axis label
plt.legend()

plt.show()
```

## Scatter Plot

```python
x = np.arange(0,10)
y1 = np.random.randint(0,100,10)
y2 = np.random.randint(0,100, 10)
```

### Creating Scatter Plots with plt.plot()

```python
#Available Markers: 'o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd'
plt.figure(figsize=(20,7.5))
plt.plot(x,y1, "rx", markersize=15) #Marker in place on line format (scatter plot)
plt.plot(x,y2, "bo-") #marker on top of line format and color (line plot with scatters)
```

### Creating Simple Scatter Plots

```python
#Available Markers: 'o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd'; #Available colors: rgbcmyk
plt.figure(figsize=(20,7.5))
plt.scatter(x,y1, s=50)
plt.scatter(x,y2, color="r", marker="x", s=50)
```

### Render Sizes, Colors and More Adjustments

```python
x = np.arange(0,10)
y1 = np.random.randn(10)
y2 = np.random.randn(10)
y3 = np.random.randint(100,10000, 10)
```

```python
plt.figure(figsize=(20,7.5))

#Available colormaps: SEQUENTIAL: 'viridis', 'plasma', 'inferno', 'magma', 'cividis'; DIVERGENT: 'PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu','RdYlBu', 'RdYlGn'; QUALITATIVE: 'flag', 'prism', 'ocean', 'gist_earth', 'terrain', 'gist_stern','gnuplot', 'rainbow', 'jet'
plt.scatter(x,y1, s=y3, c=y2, cmap="RdYlGn",  alpha=0.5)
plt.colorbar(cmap="RdYlGn", label="y2")
```

## Histograms

```python
y1 = np.random.randint(0,100,100) #classA
y2 = np.random.randint(0,100,100) #classB

plt.figure(figsize=(20,5))

#Available histtypes: 'bar', 'barstacked', 'step', 'stepfilled';
plt.subplot(1,3,1)
plt.hist(y1, histtype="stepfilled", color="red", alpha=0.5, label="classA")
plt.hist(y2, histtype="stepfilled", color="blue", alpha=0.5, label="classB")
plt.title("Combined Class Results")
plt.legend(loc="best")  #Legend locations: upper, lower, center/right, left, center e.g "best", upper left", "right", "center"

plt.subplot(1,3,2)
plt.hist(y1, bins = 5, histtype="bar", color="red")
plt.title("Class A")


plt.subplot(1,3,3)
plt.hist(y2, bins = 20, orientation="horizontal", color="blue")
plt.title("Class B")
```

## Bar Graphs

```python
x = ["A", "B", "C", "D", "E"]
y = np.random.randint(0,20, 5)
z = np.random.randint(0,20, 5)

plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
plt.bar(x,y, width=0.8) #adjusting color and width of bar
plt.title("Simple Bar Graph")

plt.subplot(1,3,2)
plt.barh(x,z, height=1, color="blue") #horizontal bar plot, width becomes height
plt.title("Horizontal Bar Graph")

plt.subplot(1,3,3)
plt.bar(x,y, color="orange", label="men")
plt.bar(x,z, color="purple", label="women")
plt.title("Stacked Bar Graph")
plt.legend(loc="best")
```

## Area Plots

```python
x = np.linspace(-np.pi, np.pi, 100)
y1 = cosine = np.cos(x)
y2 = sine = np.sin(x)

plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
plt.fill_between(x,y2)
plt.title("Sine (Fill Between Line and X-Axis)")

plt.subplot(1,3,2)
plt.fill_between(x,y1,1, color="orange")
plt.title("Cosine (Fill Between Line and X=1)")

plt.subplot(1,3,3)
plt.plot(x,y1, "k--", label="cosine")
plt.plot(x,y2, "k-", label="sine")

plt.fill_between(x,y1,y2, where=(y1>y2), label="cos > sin") #fill between y1 and y2 curve
plt.fill_between(x,y1,y2, where=(y1<y2), color="orange", label="sin > cos") #selectively fill using the where argument

plt.legend(loc="best")
plt.title("Cosine versus Sine")
```

# Advanced Visualizations/Graphs

## Contour Plots (3D Functions – Density)

```python
#Random Data for Contour Plotting
x = np.linspace(0,7.5,100)
y = np.linspace(0,7.5,100)
xx, yy = np.meshgrid(x,y) #create grid data from x and y variables
z = np.cos(yy) + np.sin(xx)
```

```python
#BMI Random Data
w = np.random.randint(25,120,100) #in kilograms
h = np.random.randint(100,200,100)/100 #in metres
ww,hh = np.meshgrid(w,h)
BMI = ww/(hh**2)
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
plt.contour(xx,yy,z) #x variable, y variable, third-dimension variable
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Contour Plot (cosy + sinx)")

plt.subplot(1,3,2)
plt.contour(xx,yy,z,30, cmap="RdBu") #increase number of intervals, add a color map
plt.title("Colored Contour Plot (cosy + sinx)")
plt.colorbar(label="z (cosy + sinx)")
plt.xlabel("x")
plt.ylabel("y")

plt.subplot(1,3,3)
plt.grid(which="major")
plt.contour(ww,hh,BMI, cmap="plasma", alpha=0.5)
plt.xlabel("Weight (kg)")
plt.ylabel("Height(m)")
plt.title("BMI Contour Plot")
plt.colorbar(label="BMI")

plt.show()
```

## 2D Histograms and Hexbins (Density)

```python
x = np.random.normal(55,5, 100)
y = np.random.normal(60,3, 100)

plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
plt.hist2d(x,y)
plt.colorbar(label="count")
plt.title("Simple 2D Histogram")

plt.subplot(1,3,2)
plt.hist2d(x,y, bins=20, cmap="Reds")
plt.colorbar(label="count")
plt.title("2D Histogram with Colormap")

plt.subplot(1,3,3)
plt.hexbin(x,y, gridsize=10, cmap="Blues")
plt.colorbar(label="count")
plt.title("Hexbin with Colormap")
```

## Error Bars

```python
x = np.arange(100)
y1 = np.random.normal(60,3, 100)
y2 = np.random.normal(55,5, 100)
dy1 = 0.5
dy2 = 0.5

plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
plt.errorbar(x,y1, yerr=dy1)
plt.title("Simple Error Bar (Line)")

plt.subplot(1,3,2)
plt.errorbar(x,y2, yerr=dy2, fmt=".", color="Blue")
plt.title("Simple Error Bar (Scatter)")

plt.subplot(1,3,3)
plt.errorbar(y1,y2, xerr=dy1, yerr=dy2, fmt="o", color="Green")
plt.title("X-Y Errorbars (Scatter)")
```
